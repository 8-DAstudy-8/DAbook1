1. 데이터를 수집한 후 불러와서 확인.

```python
df = pd.read_csv(os.path.join(root_dir, '2001_Dataset.csv'),encoding='cp949')
df.head()
```

2. 태어난 날짜의 최고기온과 최저기온을 나타내보기
