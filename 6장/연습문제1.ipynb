###'품종' 열을 기준으로 최댓값과 최솟값을 출력

agg() 메서드에는 max, min. mean, sum, std, var 등 다양한 집계 함수를 사용할 수 있음.

```python
iris.groupby('Species').agg('max')
```

```python
iris.groupby('Species').agg('min')
```
