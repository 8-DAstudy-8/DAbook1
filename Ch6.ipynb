```python
import pandas as pd
titanic=pd.read_csv('train.csv')
titanic.head()
```

```python
titanic.sort_values('Fare',ascending=False)
```

```python
pd.DataFrame(titanic['Survived'].value_counts())
```

```python
titanic.isnull().sum()
```

```python
#       
titanic.drop(['Cabin'],axis=1,inplace=True)
titanic.columns
```

```python
#2         .
#   
titanic['Embarked'].value_counts()
```

```python
titanic['Embarked']=titanic['Embarked'].fillna('S')
titanic.isnull().sum()
```

```python
#       
avg=titanic['Age'].mean()
titanic['Age']=titanic['Age'].fillna(avg)
titanic.isnull().sum()
```

```python
titanic['Pclass'].groupby(titanic["Embarked"]).mean()
```

```python
titanic.groupby(["Pclass", "Sex"]).mean(numeric_only=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Pclass', hue='Survived', data=titanic)
plt.show()
```
