### (2) 데이터 병합하기

```python
city = list(happy_life['시도'].unique())
happy_merge = DataFrame( {'시도' : city} )
happy_merge
```

```python
life = happy_life['삶의 만족도'].groupby(by=happy_life['시도']).mean()
happy_merge = merge(happy_merge, life, on='시도')
happy_merge.head()
```

```python
health = happy_health['평균'].groupby(by=happy_health['시도']).mean()
happy_merge = merge(happy_merge, health.rename('건강'), on='시도')
happy_merge.head()
```

```python
safe = happy_safe['평균'].groupby(by=happy_safe['시도']).mean()
happy_merge = merge(happy_merge, safe.rename('안전'), on='시도')
happy_merge.head()
```

```python
environ = happy_environ['평균'].groupby(by=happy_environ['시도']).mean()
happy_merge = merge(happy_merge, environ.rename('환경'), on='시도')
happy_merge.head()
```

```python
econo = happy_econo['평균'].groupby(by=happy_econo['시도']).mean()
happy_merge = merge(happy_merge, econo.rename('경제'), on='시도')
happy_merge.head()
```

```python
edu = happy_edu['평균'].groupby(by=happy_edu['시도']).mean()
happy_merge = merge(happy_merge, edu.rename('교육'), on='시도')
happy_merge.head()
```

```python
relation = happy_relation['평균'].groupby(by=happy_relation['시도']).mean()
happy_merge = merge(happy_merge, relation.rename('관계및사회참여'), on='시도')
happy_merge.head()
```

```python
leisure = happy_leisure['평균'].groupby(by=happy_leisure['시도']).mean()
happy_merge = merge(happy_merge, leisure.rename('여가'), on='시도')
happy_merge
```

```python
seoul_data = happy_life[happy_life['시도'] == '서울특별시']
gu=list(seoul_data['구군'].unique())
gu_merge=DataFrame({'구군':gu})
gu_merge
```

```python
gu_merge = gu_merge.drop(columns=['삶의 만족도'])
gu_merge
```

```python
life = seoul_data.groupby('구군')['삶의 만족도'].mean().reset_index()
gu_merge = pd.merge(gu_merge, life, on='구군')
gu_merge
```

```python
seoul_data2 = happy_health[happy_health['시도'] == '서울특별시']
health = seoul_data2.groupby('구군')['평균'].mean().reset_index()
health = health.rename(columns={'평균': '건강'})
gu_merge = pd.merge(gu_merge, health, on='구군', how='left')

gu_merge
```

```python
seoul_data3 = happy_safe[happy_safe['시도'] == '서울특별시']
safe = seoul_data3.groupby('구군')['평균'].mean().reset_index()
safe = safe.rename(columns={'평균': '안전'})
gu_merge = pd.merge(gu_merge, safe, on='구군', how='left')
gu_merge
```

```python
seoul_data4 = happy_environ[happy_environ['시도'] == '서울특별시']
environ = seoul_data4.groupby('구군')['평균'].mean().reset_index()
environ = environ.rename(columns={'평균': '환경'})
gu_merge = pd.merge(gu_merge, environ, on='구군', how='left')
gu_merge
```

```python
seoul_data5 = happy_econo[happy_econo['시도'] == '서울특별시']
econo = seoul_data5.groupby('구군')['평균'].mean().reset_index()
econo = econo.rename(columns={'평균': '경제'})
gu_merge = pd.merge(gu_merge, econo, on='구군', how='left')
gu_merge
```

```python
seoul_data6 = happy_edu[happy_edu['시도'] == '서울특별시']
edu = seoul_data6.groupby('구군')['평균'].mean().reset_index()
edu = edu.rename(columns={'평균': '교육'})
gu_merge = pd.merge(gu_merge, edu, on='구군', how='left')
gu_merge
```

```python
seoul_data7 = happy_relation[happy_safe['시도'] == '서울특별시']
relation = seoul_data7.groupby('구군')['평균'].mean().reset_index()
relation = relation.rename(columns={'평균': '관계및사회참여'})
gu_merge = pd.merge(gu_merge, relation, on='구군', how='left')
gu_merge
```

```python
seoul_data8 = happy_leisure[happy_safe['시도'] == '서울특별시']
leisure = seoul_data8.groupby('구군')['평균'].mean().reset_index()
leisure = leisure.rename(columns={'평균': '여가'})
gu_merge = pd.merge(gu_merge, leisure, on='구군', how='left')
gu_merge
```

```python
corr_matrix = gu_merge.select_dtypes(include=['float64', 'int64']).corr()

print(corr_matrix)
```
