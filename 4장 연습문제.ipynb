##3번 문제

```python
import numpy as np
arr = np.arange(1,21)
print('마지막 원소 : ', arr[-1])
print('모든 원소의 합 : ', arr.sum())
print('모든 원소의 평균 : ', arr.sum()/arr.size)
```

```python
import numpy as np
num = np.random.randint(1,31)
while True:
  user_input = int(input("숫자 입력 (종료 0 ): "))
  if user_input ==num:
    print("정답!!")
    break
  elif user_input>=num:     #python에는 else if 대신 elif
    print("더 낮은 숫자 입력!")
    continue
  elif user_input <= num:
    print("더 높은 숫자 입력!")
    continue
  elif user_input == 0:
    break
    # 1~30이 아닌 숫자 입력햇을때 try except 문으로 해주기 (python은 try catch문이 아니라 except라고 함)
```

##4번 문제

```python
import numpy as np
salary = [200,150,180,220,250]
new_salary=np.array(salary,dtype = 'float64')*1.5
print('인상된 금액 : ',new_salary)
avg_salary = new_salary.sum()/new_salary.size
print('평균 금액 : ',avg_salary)
result = np.array(new_salary>=avg_salary)
print('평균 이상인 원소 : ', new_salary[result])
print('평균 이상인 원소의 개수 : ', new_salary[result].size) #result는 true, false의 값을 가지고 있음. 즉 요소가 5개임. 그러므로 result.size하면 안됨
```
