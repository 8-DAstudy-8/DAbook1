5) 판다스 결측데이터 처리
- 판다스는 누락된 데이터를 표시할 때 NaN으로 표시하고 연산에는 포함되지 않는다.

```python
#결측데이터 확인
print(df[df.mean_wind.isna()])#isna()함수는 값이 nan일때 True, 그렇지 않으면 False가 된다.
print(df[df.mean_wind.isnull()])#isnull도 가능하다.
```

```python
#결측데이터 개수 확인:컬럼별 결측데이터 개수 확인
print(df.isnull().sum())#isnull()함수가 결측데이터이면 True값을 반환하고, 유효한 데이터가 존재하면 false를 반환한다. notnull()은 반대
print()
print(df.isna().sum())
```

- 결측데이터 삭제하기
  - dropna()함수 이용하여 삭제
  - 매개변수
    - axis: 축을 행 또는 열로 결정한다, 0,index-> 누락값 포함 행 삭제 1,columns-> 누락값포함 열 삭제. 기본적으로 0이다.
    - how: any->null값이 있는경우 행 또는 열 삭제한다. all->모든 값이 누락된 경우 행 또는 열을 삭제
    - inplace: True로 설정하면 호출자 DataFrame을 변경하는 부울 값 기본적으로는 False

```python
#결측데이터가 있는 행 삭제 후 확인
df2=df.dropna()
df2.isnull().sum()
```

```python
#결측데이터 대체:평균값으로 대체
df.index=df['date']
df.drop('date',axis=1,inplace=True)
df.fillna(df.mean(),inplace=True)
df.isnull().sum()
```

7) 판다스 데이터 그룹핑
- 데이터 그룹핑은 데이터를 특정한 값에 기반에 묶는 기능으로 groupby()함수를 사용해 통계량을 요약한다.
- 통계량 관련 함수에는 mean,std(표준편차),var(분산),max,min,mode(최빈값) 등이 있다.

```python
#한열을 기준으로 그룹화
s1=pd.read_excel('nation.xlsx')
s2=s1.drop(['성별'],axis=1)
s2.groupby('국적코드').sum()
```

```python
#여러열을 기준으로 그룹화->여러열을 그룹으로 묶어서 계층적인 인덱스를 구성한다.
s1=pd.read_excel('nation.xlsx')
s1.groupby(['국적코드','성별']).sum()
```
